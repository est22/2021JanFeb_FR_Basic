# 1. 데이터프레임의 구조
## 1-1. 데이터 내용 미리보기 : head(), tail()
## 1-2. 데이터 요약 정보 확인하기 : shape(), info(), dtypes, describe()

```python
import pandas as pd

df = pd.read_csv('./source/part3/auto-mpg.csv',header=None)

df.columns = ['mpg','cylinders','displacement','horsepower','weight',
              'acceleration','model year','origin','name']

# 데이터프레임 df의 내용을 일부 확인 
print(df.head()) 
print('\n')
print(df.tail()) 

# df의 모양과 크기 확인: (행의 개수, 열의 개수)를 투플로 반환 
print(df.shape)
print('\n')

# 데이터프레임 df의 내용 확인 
print(df.info())
print('\n')

# 데이터프레임 df의 자료형 확인 
print(df.dtypes)
print('\n')

# 시리즈(mpg 열)의 자료형 확인 
print(df.mpg.dtypes)
print('\n')

# 데이터프레임 df의 기술통계 정보 확인 
print(df.describe())
print('\n')
print(df.describe(include='all'))
```

## 1-3. 데이터 개수 확인
### 각 열의 데이터 개수 : count()

```python
# df의 각 열이 가지고 있는 원소 개수 확인
print(df.count()) # 좌:열 이름, 우:각 열의 데이터 개수
print('\n')

# df.count()가 반환하는 객체 타입 출력
print(type(df.count())) 
```

### 각 열의 고유값 개수 : value_counts()

```python
# 각 열의 고유값 개수
unique_values = df['origin'].value_counts() # origin 열의 고유값 종류와 개수
print(unique_values)
print('\n')

# value_counts 메소드가 반환하는 객체 타입 출력
print(type(unique_values))
```

# 2. 통계 함수 적용
## 2-1. 평균값 : mean()





```python
# 평균값 
print(df.mean())
print('\n')
print(df['mpg'].mean())
print(df.mpg.mean())
print('\n')
print(df[['mpg','weight']].mean())
```

## 2-2. 중간값 : median()

```python
# 중간값 
print(df.median())
print('\n')
print(df['mpg'].median())
```

## 2-3. 최대값 : max()

```python
# 최대값 
print(df.max())
print('\n')
print(df['mpg'].max())
```

## 2-4. 최소값 : min()

```python
# 최소값 
print(df.min())
print('\n')
print(df['mpg'].min())
```

## 2-5. 표준편차 : std()

```python
# 표준편차 
print(df.std())
print('\n')
print(df['mpg'].std())
```

## 2-6. 상관계수 : corr()

```python
# 상관계수 
print(df.corr())
print('\n')
print(df[['mpg','weight']].corr())
```

# 3. 판다스 내장 그래프 도구 활용
## * 선 그래프

```python
import pandas as pd

df = pd.read_excel('./source/part3/남북한발전전력량.xlsx',header=0)

df_ns = df.iloc[[0,5],2:]
#print(df_ns)
df_ns.index = ['South','North'] # 행 인덱스 변경
df_ns.columns = df_ns.columns.map(int) # 열 이름의 자료형을 정수형으로 변경
print(df_ns.head())
print('\n')

```

```python
df_ns=df_ns.astype(int)
df_ns.plot()
# 행, 열 전치하여 막대 그래프 그리기
tdf_ns = df_ns.T
print(tdf_ns.head())
print('\n')
tdf_ns.plot()
```

## * 막대 그래프

```python
tdf_ns.plot(kind='bar')
```

## * 히스토그램

```python
# 행, 열 전치하여 히스토그램 그리기
tdf_ns = df_ns.T
tdf_ns.plot(kind='hist')
```
